```python
import os
import pandas as pd
import snowflake.connector

# -----------------------------
# Config
# -----------------------------
CSV_FILE = r"C:\Users\harsh\OneDrive\Desktop\Sales_Analysis\raw_data_cleaned_snowflake.csv"
SNOWFLAKE_CONFIG = {
    "user": "SHREEHARSHA",
    "password": "Smhk@07072002k",
    "account": "OGJVEXN-HI94808",
    "warehouse": "MY_WH",
    "database": "SALES_PROJECT",
    "schema": "CLEANED_DATA"
}
TABLE_NAME = "SALES_DATA"
UNIQUE_KEY = "ORDER_ID"

# -----------------------------
# Helper Functions
# -----------------------------
def get_max_order_id(cur):
    cur.execute(f"SELECT MAX({UNIQUE_KEY}) FROM {TABLE_NAME}")
    result = cur.fetchone()[0]
    return result if result is not None else 0

def merge_new_rows(df_new, cur,conn):
    if df_new.empty:
        print("‚úÖ No new rows to insert.")
        return

    temp_csv = "temp_new_rows.csv"
    df_new.to_csv(temp_csv, index=False)

    cur.execute("CREATE STAGE IF NOT EXISTS MY_STAGE;")
    cur.execute(f"PUT file://{os.path.abspath(temp_csv)} @MY_STAGE AUTO_COMPRESS=FALSE;")


    merge_sql = f"""
    MERGE INTO {TABLE_NAME} AS target
    USING (
        SELECT 
            $1 AS ORDER_ID,
            TO_DATE($2, 'DD-MM-YYYY') AS ORDER_DATE,
            $3 AS CUSTOMER_ID,
            $4 AS CUSTOMER_NAME,
            $5 AS REGION,
            $6 AS PRODUCT,
            $7 AS CATEGORY,
            $8 AS QUANTITY,
            $9 AS UNIT_PRICE,
            $10 AS DISCOUNT,
            $11 AS REVENUE
        FROM @MY_STAGE/{os.path.basename(temp_csv)}
        (FILE_FORMAT => 'my_csv_format')
    ) AS source
    ON target.ORDER_ID = source.ORDER_ID
    WHEN MATCHED THEN UPDATE SET
        target.ORDER_DATE = source.ORDER_DATE,
        target.CUSTOMER_ID = source.CUSTOMER_ID,
        target.CUSTOMER_NAME = source.CUSTOMER_NAME,
        target.REGION = source.REGION,
        target.PRODUCT = source.PRODUCT,
        target.CATEGORY = source.CATEGORY,
        target.QUANTITY = source.QUANTITY,
        target.UNIT_PRICE = source.UNIT_PRICE,
        target.DISCOUNT = source.DISCOUNT,
        target.REVENUE = source.REVENUE
    WHEN NOT MATCHED THEN INSERT (
        ORDER_ID, ORDER_DATE, CUSTOMER_ID, CUSTOMER_NAME, REGION,
        PRODUCT, CATEGORY, QUANTITY, UNIT_PRICE, DISCOUNT, REVENUE
    ) VALUES (
        source.ORDER_ID, source.ORDER_DATE, source.CUSTOMER_ID, source.CUSTOMER_NAME,
        source.REGION, source.PRODUCT, source.CATEGORY, source.QUANTITY,
        source.UNIT_PRICE, source.DISCOUNT, source.REVENUE
    );
    """

    cur.execute(merge_sql)
    conn.commit()
    os.remove(temp_csv)
    print(f"üöÄ Upserted {len(df_new)} rows into {TABLE_NAME}.")
                    
    cur.execute(merge_sql)
    if os.path.exists(temp_csv):
       os.remove(temp_csv)
    else:
       print(f"‚ö†Ô∏è File not found for cleanup: {temp_csv}")
    print(f"üöÄ Upserted {len(df_new)} rows into {TABLE_NAME}.")

# -----------------------------
# Main Execution
# -----------------------------
print("üöÄ Starting Snowflake CDC Sync...")

try:
    conn = snowflake.connector.connect(**SNOWFLAKE_CONFIG)
    cur = conn.cursor()

    df = pd.read_csv(CSV_FILE)
    max_id = get_max_order_id(cur)
    df_new = df[df[UNIQUE_KEY] > max_id]

    merge_new_rows(df_new, cur,conn)

    cur.close()
    conn.close()
    print("‚úÖ CDC sync completed successfully.")

except Exception as e:
    print("‚ùå Error during CDC sync:", e)
```
